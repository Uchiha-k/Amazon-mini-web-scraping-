```python
import requests
from bs4 import BeautifulSoup


def get_amazon_product(url):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0 Safari/537.36",
        "Accept-Language": "en-US,en;q=0.9",
    }

    page = requests.get(url, headers=headers)
    soup = BeautifulSoup(page.content, "html.parser")

    # Title
    title_tag = soup.find("span", id="productTitle")
    title = title_tag.get_text(strip=True) if title_tag else None

    #Price
    whole = soup.find("span", class_= "a-price-whole")
    fraction = soup.find("span", class_ = "a-price-fraction")

    if whole and fraction:
        price = whole.get_text(strip=True).replace(",", "").replace(".", "")
        fraction = fraction.get_text(strip=True)
        price = f"{price}.{fraction}"
    else:
        price = None

    return {
        "title": title,
        "price": price
    }

#RUN
url = "https://www.amazon.com/dp/B07PJL8NZ3"
product = get_amazon_product(url)
print(product)
```

```python
import csv

#url
url = "https://www.amazon.com/dp/B07PJL8NZ3"

#get info on product
product = get_amazon_product(url)

#access values
title = product['title']
price = product['price']

# Write to CSV
with open("revision.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["Title", "Price"])  # header
    writer.writerow([title, price])
```
